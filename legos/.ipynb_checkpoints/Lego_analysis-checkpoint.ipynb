```python
# importing necessary libraries
import pandas as pd  # for data manupilation
from matplotlib import pyplot as plt # for visualization
import seaborn as sns # for visualization
import numpy as np
```

```python
# import data
df=pd.read_csv("datasets/lego_sets.csv")
parent_themes=pd.read_csv("datasets/parent_themes.csv")
```

```python
df.head(10)
```

<h1>May the force be with YOU!!!!</h1>
<h2>What percentage of all licensed sets ever released were Star Wars themes?<h2>

```python
# merge the data frames
merged_df=df.merge(parent_themes,left_on="parent_theme",right_on="name")
merged_df.dropna(subset=["set_num"],inplace=True)

filter1=(merged_df["is_licensed"]==True)
liscensed_df=merged_df[filter1]

filter2=(merged_df["is_licensed"]==True)
unliscensed_df=merged_df[filter2]
```

```python
count_licensed=liscensed_df.groupby("parent_theme",as_index=False)["is_licensed"].count()
count_licensed.rename(columns={"is_licensed":"num_licensed"},inplace=True)
count_licensed.sort_values(by="num_licensed",ascending=False,inplace=True)
```

```python
# count_licensed.loc[count_licensed["parent_theme"]=="Star Wars","num_licensed"]/count_licensed["num_licensed"].sum()
count_licensed=count_licensed.assign(market_share_pct=lambda x: x.num_licensed/count_licensed["num_licensed"].sum()*100)
```

```python
count_licensed
```

```python
fig,ax=plt.subplots(figsize=(15,6))
labels=count_licensed["parent_theme"]
# define a function to label the percentage
def autolabel(barplot):
    """
    Attach a text label above each bar displaying its height
    """
    for bar in barplot:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., 1.05*height,
                "{:.2f}%".format(height),
                ha='center', va='bottom')



bar=ax.bar(labels,count_licensed["market_share_pct"])
plt.xticks(rotation=60)
plt.grid(axis="y")
autolabel(bar)
plt.show()
```

<h1>New Era</h1>
<h2>In which year was Star Wars not the most popular liscened theme<strong>(Year which Star Wars get beat out by other set)</strong></h2>

```python
yearly_df=liscensed_df.groupby(["year","parent_theme"],as_index=False).sum()
yearly_df.sort_values("is_licensed",ascending=False,inplace=True)
max_yearly_df=yearly_df.drop_duplicates(["year"])
max_yearly_df.sort_values(by=["year"],inplace=True)
max_yearly_df
```

<p1><u>Star Wars was the beast through out the years but it get beat out in 2017 by super heros lego sets</u></p1>

```python
fig,ax=plt.subplots(figsize=(15,6))
#define a function to draw a text on the  bar plot
def labeling(barplot):
    """
    Attach a text label above each bar displaying its height and Name
    """
    for bar,name in zip(barplot,max_yearly_df["parent_theme"]):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., 1.05*height,
                "{:.0f}".format(height),
                ha='center', va='bottom')
        ax.text(bar.get_x()+bar.get_width()/2.,.85*height,
               name,
               ha="center",va="top",fontfamily="fantasy",rotation="vertical",
                alpha=.9,color="yellow",fontsize="large")


bar2=ax.bar(max_yearly_df["year"].astype("str"),max_yearly_df["is_licensed"])
labeling(bar2)
plt.show()
```

<h1>Number of sets by years</h1>
<h2>How many unieque sets were released each year</h2>

```python
# first thing first check if there is any missing data set_num column
merged_df.set_num.isnull().any()
```

```python
count_unique_sets=merged_df.groupby("year",as_index=False)["set_num"].nunique()
count_unique_sets.rename(columns={"set_num":"num_unique_sets"},inplace=True)
count_unique_sets
```

```python
fig,ax=plt.subplots(figsize=(15,6))
sns.lineplot(count_unique_sets["year"],count_unique_sets["num_unique_sets"],ax=ax)
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
ymax=max(count_unique_sets["num_unique_sets"])
xpos=np.where(count_unique_sets["num_unique_sets"]==ymax)
xmax=count_unique_sets.to_numpy()[xpos][0][0]
ax.grid(axis="x")
ax.annotate('year 2014', xy=(xmax, ymax), xytext=(xmax-10, ymax - 5), arrowprops=dict(facecolor='black'))

```

<p><strong>2014</strong> was the year with the most unique legos set</p>
